# Train data preprocessing

```python
df1 = data1.copy()
df1 = df1.drop(['PassengerId'], axis = 1)
df1 = df1.drop(['Name'], axis = 1)
df1 = df1.drop(['Cabin'], axis = 1)
df1 = df1.drop(['Ticket'], axis =1)
df1['Sex'] = df1['Sex'].map({'male': 1, 'female': 0})
df1.head()
```

## Age

```python
df1Age = df1.copy()
ageArray = []
ageMean = df1Age['Age'].mean()
for i in df1Age['Age']:
    if pd.isna(i):
        i = ageMean
        ageArray.append(i)
    else:
        ageArray.append(i)

dfAgeArray = pd.DataFrame(ageArray)
dfAgeArray
df1Age = pd.concat([df1Age, dfAgeArray], axis = 1)
df1Age = df1Age.drop(['Age'], axis = 1)
df1Age.head()
```

```python
df1Age.columns.values
newColumns = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Age']
df1Age.columns = newColumns
df1Age.head()
```

## Pclass

```python
df1AgePClass = df1Age.copy()
df1AgePClass.head()
```

```python
pclassColumns = pd.get_dummies(df1AgePClass['Pclass'], drop_first=True)
df1AgePClass = pd.concat([df1AgePClass, pclassColumns], axis =1)
df1AgePClass.head()
```

```python
df1AgePClass.columns.values
newColumns = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Fare', 'Embarked',
       'Age', 'Pclass 1', 'Pclass 2']
df1AgePClass.columns = newColumns
df1AgePClass = df1AgePClass.drop(['Pclass'], axis = 1)
df1AgePClass.head()
```

## SibSp

```python
df1AgePClassSibSp = df1AgePClass.copy()
df1AgePClassSibSp.head()
df1AgePClassSibSp.info()
```

## Parch

```python
df1AgePClassSibSpParch = df1AgePClassSibSp.copy()
df1AgePClassSibSpParch.head()
```

## Fare

```python
df1AgePClassSibSpParchFare = df1AgePClassSibSpParch.copy()
df1AgePClassSibSpParchFare.info()
```

## Embarked

```python
df1AgePClassSibSpParchFareEmbarked = df1AgePClassSibSpParchFare.copy()
df1AgePClassSibSpParchFareEmbarked['Embarked'].unique()
df1AgePClassSibSpParchFareEmbarked['Embarked'] = df1AgePClassSibSpParchFareEmbarked['Embarked'].map({'S': int(1), 'C': int(2), 'Q': int(3)})
df1AgePClassSibSpParchFareEmbarked.head()

embarkedArray = []
for i in df1AgePClassSibSpParchFareEmbarked['Embarked']:
    if pd.isna(i):
        i = int(2)
        embarkedArray.append(i)
    else:
        embarkedArray.append(i)
        
dfEmbarkedArray = pd.DataFrame(embarkedArray)
dfEmbarkedArray
df1AgePClassSibSpParchFareEmbarked = pd.concat([df1AgePClassSibSpParchFareEmbarked, dfEmbarkedArray], axis = 1)
#df1AgePClassSibSpParchFareEmbarked.info()
df1AgePClassSibSpParchFareEmbarked = df1AgePClassSibSpParchFareEmbarked.drop(['Embarked'], axis = 1)
df1AgePClassSibSpParchFareEmbarked.head()

embarkedColumns = pd.get_dummies(df1AgePClassSibSpParchFareEmbarked[0], drop_first=True)
df1AgePClassSibSpParchFareEmbarked = pd.concat([df1AgePClassSibSpParchFareEmbarked, embarkedColumns], axis =1)
df1AgePClassSibSpParchFareEmbarked = df1AgePClassSibSpParchFareEmbarked.drop([0], axis = 1)

df1AgePClassSibSpParchFareEmbarked.columns.values
newColumns = ['Survived', 'Sex', 'SibSp', 'Parch', 'Fare', 'Age', 'Pclass 1',
       'Pclass 2', 'Embarked C', 'Embarked D']
df1AgePClassSibSpParchFareEmbarked.columns = newColumns

df1AgePClassSibSpParchFareEmbarked.head()
```

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df1Scaled = df1AgePClassSibSpParchFareEmbarked.copy()
df1Scaled[['SibSp', 'Parch', 'Fare', 'Age']] = scaler.fit_transform(df1Scaled[['SibSp', 'Parch', 'Fare', 'Age']])
df1Scaled.head()
```

# Saving data to npz file

## Spliting the data

```python
dfTrain = df1Scaled.copy()
dfTest = dfTestScaled.copy()
```

```python
samples_count = dfTrain.shape[0]
train_samples_count = int(0.8*samples_count)
validation_samples_count = int(0.1*samples_count)
test_samples_count = samples_count - train_samples_count - validation_samples_count
```

```python
targetsAll = dfTrain.iloc[:,0]
```

```python
dfTrain = dfTrain.drop(['Survived'], axis = 1)
```

## Spliting the train data

```python
trainInputs = dfTrain[:train_samples_count]
trainTargets = targetsAll[:train_samples_count]

validationInputs = dfTrain[train_samples_count:train_samples_count + validation_samples_count]
validationTargets = targetsAll[train_samples_count:train_samples_count + validation_samples_count]

testInputs = dfTrain[train_samples_count + validation_samples_count:]
testTargets = targetsAll[train_samples_count + validation_samples_count:]
```

```python
np.savez('titanicTrainData', inputs = trainInputs, targets = trainTargets)
np.savez('titanicValidationData', inputs = validationInputs, targets = validationTargets)
np.savez('titanicTestData', inputs = testInputs, targets = testTargets)
```
